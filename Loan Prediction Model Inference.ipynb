```python
# import libraries
import pandas as pd
import numpy as np
import pickle
import random
```

```python
with open('full_process.pkl', 'rb') as file:
  full_process = pickle.load(file)
```

```python
# Pengaturan Seed
seed_value = 10
random.seed(seed_value)
np.random.seed(seed_value)

# Jumlah data point yang akan dibuat
n = 20


Married = np.random.choice(['1', '0'], 20)
Education = np.random.choice(['1', '0'], 20)
Credit_History = np.random.randint(0, 2, 20)
Property_Area = np.random.choice(['0', '1', '2'], 20)

# Membuat DataFrame
data_inf = pd.DataFrame({
    'Married' : Married,
    'Education': Education,
    'Credit_History': Credit_History,
    'Property_Area': Property_Area,

    
})

data_inf
```

```python
y_pred_inf = full_process.predict(data_inf)
y_pred_inf
```

```python
# Mengkonversi ke dataframe

y_pred_inf_df = pd.DataFrame(y_pred_inf, columns=['Target - Prediction'])
y_pred_inf_df
```

```python
# Menyatukan data_inf dengan y_pred_inf_df

tabel_inference = pd.concat([data_inf, y_pred_inf_df], axis=1)
tabel_inference
```
